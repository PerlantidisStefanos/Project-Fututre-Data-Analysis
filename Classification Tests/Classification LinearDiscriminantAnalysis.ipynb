Feature Selection

```python
X.shape
```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectFromModel
lsvc = LinearSVC(C=0.01, penalty="l1", dual=False).fit(X, y)
model = SelectFromModel(lsvc, prefit=True)
X_new = model.transform(X)
X_new.shape
```

```python
X_tr, X_ts, y_tr, y_ts = train_test_split(X_new, y, test_size=0.2, stratify=y)
X_ts.shape
```

Training Model

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda= LinearDiscriminantAnalysis()
```

```python
grid_lda= { 'solver' : ['svd', 'lsqr', 'eigen'],
            'n_components': [20, 50, 100,200],
            'tol' : [0.00000001, 0.0000001, 0.000001],
           }
clf_lda= GridSearchCV(lda, grid_lda, cv=3, scoring='recall')
clf_lda.fit(X_tr, y_tr)

lda_preds = clf_lda.predict(X_ts)
from sklearn.metrics import classification_report , accuracy_score , precision_recall_fscore_support
print(clf_lda.best_params_, '\n')
print('accuracy: {:.2f}%'.format(accuracy_score(y_ts, lda_preds)*100))

print(classification_report(y_ts, lda_preds))
confusion_matrix(y_ts,lda_preds)
```

```python
grid_lda2= { 'solver' : ['svd'],
            'n_components': [6, 8, 10, 12, 14],
            'tol' : [0.0000000001, 0.000000001, 0.00000001],
           }
clf_lda2= GridSearchCV(lda, grid_lda2, cv=5, scoring='recall')
clf_lda2.fit(X_tr, y_tr)

lda_preds2 = clf_lda2.predict(X_ts)
from sklearn.metrics import classification_report , accuracy_score , precision_recall_fscore_support
print(clf_lda2.best_params_, '\n')
print('accuracy: {:.2f}%'.format(accuracy_score(y_ts, lda_preds2)*100))

print(classification_report(y_ts, lda_preds2))
confusion_matrix(y_ts,lda_preds2)
```
